### This project is using nltk, numpy, wordcloud, matplotlib

### Imports:

```python
#import jieba
import nltk
import numpy as np
from PIL import Image

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

### Read the text file:
+ *Harry Potter_1* (txt file)


```python
path = "Harry_Potter_1_Sorcerer's_Stone.txt"
file = open(path,"r",encoding="utf-8")
text=file.read()
file.close()
```

### Set a picture that you want to 'mask' (shape) your word cloud:

```python
hp_mask = np.array(Image.open("img/gryffindor.webp"))
```

### Main Function:
+ create wordcloud
+ set color_fuction(follow by the colors of the picture)
+ save plot as picture file

```python
w = WordCloud(max_words=3000,mask=hp_mask, mode="RGBA").generate(text)
#w.to_file("img/HP_wordcloud.png")

# show

image_colors = ImageColorGenerator(hp_mask)
plt.figure(figsize=[30,30])
plt.imshow(w.recolor(color_func = image_colors), interpolation='bilinear')
plt.axis("off")

plt.savefig("img/HPwc.png", format="png")
#plt.show()
```
